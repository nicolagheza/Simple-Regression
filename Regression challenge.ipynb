```python
# Importing dependencies
import pandas as pd
import numpy as np 
from sklearn import linear_model
from sklearn import svm
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

#read data
dataframe = pd.read_csv('challenge_dataset.txt')
x_values = np.array(dataframe[['X']])
y_values = np.array(dataframe[['Y']])
#split data in training and test sets
X_train, X_test, y_train, y_test = train_test_split(x_values, y_values, test_size=0.4, random_state=0)
```

```python
linear_clf = linear_model.LinearRegression()
svm_clf = svm.SVR(kernel= 'linear', C= 1e3) # defining the support vector regression models
#train classifier
linear_clf.fit(X_train, y_train)
svm_clf.fit(X_train, y_train.ravel())
```

```python
print 'Score Linear SVM (on test set): ', svm_clf.score(X_test, y_test)
print 'Score Linear SVM (on entire set): ', svm_clf.score(x_values, y_values)
print 'Mean Absolute Error: ', mean_absolute_error(y_values, svm_clf.predict(x_values))
print 'Score Linear Reg (on test set): ', linear_clf.score(X_test, y_test)
print 'Score Linear Reg (on entire set): ', linear_clf.score(x_values, y_values)
print 'Mean Absolute Error: ', mean_absolute_error(y_values, linear_clf.predict(x_values))
plt.scatter(x_values, y_values, color='black', label='Data')
plt.plot(x_values, svm_clf.predict(x_values), color='red', label='Linear SVM Model')
plt.plot(x_values, linear_clf.predict(x_values), color='green', label='Linear Reg Model')
plt.legend()
plt.show()
```
